```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data/motDegPID.txt", names=['degB', 'degC', 'aim', 'powB', 'powC'], sep=";")
```

```python
data['res'] = data['degB'] / data['degC']
data['windowedRes'] = data['res'].rolling(10).mean()
data = data.dropna()
```

```python
x_vals = np.arange(0, len(data['degB']))
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax.plot(x_vals, data['degB'], label='degB')
ax.plot(x_vals, data['degC'], label='degC')
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax.plot(x_vals, data['powB'], label='powB')
ax.plot(x_vals, data['powC'], label='powC')
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax.plot(x_vals, data['res'], label='res')
ax.plot(x_vals, data['windowedRes'], label='windowedRes')
ax.plot(x_vals, data['aim'], label='aim')
```

```python
print(' : ', max([abs(el) for el in data['aim'] - data['res']]))
```
