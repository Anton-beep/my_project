```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data.csv", names=['seconds', 'seconds_from_start', 'motBpower', 'motCpower', 'motBEnc', 'motCEnc'])
plt.plot(data['seconds_from_start'], data['motBpower'], label='motBpower')
plt.plot(data['seconds_from_start'], data['motCpower'], label='motCpower')
plt.xlabel('seconds_from_start')
plt.legend()
```

```python
plt.plot(data['seconds_from_start'], data['motBEnc'], label='motBEnc')
plt.plot(data['seconds_from_start'], data['motCEnc'], label='motCEnc')
plt.xlabel('seconds_from_start')
plt.legend()
```

```python
SP_B = 5
SP_C = 90

print(abs((SP_B / SP_C) / (data['motBEnc'].iloc[-1] / data['motCEnc'].iloc[-1])))
if 0.95 < abs((SP_B / SP_C) / (data['motBEnc'].iloc[-1] / data['motCEnc'].iloc[-1])) < 1.05:
    print('good')
else:
    print('bad')
```

```python
print(f'max power B: {max(data["motBpower"])}, min power B: {min(data["motBpower"])}, amp: {max(data["motBpower"]) - min(data["motBpower"])}')
print(f'max power C: {max(data["motCpower"])}, min power C: {min(data["motCpower"])}, amp: {max(data["motCpower"]) - min(data["motCpower"])}')
```
